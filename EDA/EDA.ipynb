```python
# 1. Generate Summary Statistics
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv(r'C:\Users\OMEN\Desktop\Work\EDA_Task2\Data\Titanic-Dataset.csv')
```

```python
# Summary statistics
display(df.describe(include='all'))
df.info()
```

```python
# 2. Create Histograms and Boxplots
num_cols = ['Age', 'Fare', 'SibSp', 'Parch']
```

```python
for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Histogram of {col}')
    plt.show()
```

```python
for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
# 3. Feature Relationships
# Pairplot (subset for speed)
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived')
plt.show()
```

```python
# Correlation Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# 4. Identify Patterns, Trends, or Anomalies
# Example: Grouped barplot for categorical feature vs target
sns.barplot(data=df, x='Pclass', y='Survived')
plt.title('Survival Rate by Passenger Class')
plt.show()

sns.barplot(data=df, x='Sex', y='Survived')
plt.title('Survival Rate by Sex')
plt.show()
```
